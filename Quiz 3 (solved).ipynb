## Exercise 1

Draw scatterplots for each of these datasets where you can see the relation for

- economics: date, unemploy  (date and unemployment)
- diamonds: carat, price (carat and price)
- mpg: cty, hwy (city vs highway miles per gallon)

```python
economics.head()
```

```python
diamonds.head()
```

```python
mpg.head()
```

```python
(ggplot(data=economics, mapping=aes(x='date', y='unemploy'))+
geom_point())
```

```python
(ggplot(data=diamonds, mapping=aes(x='carat', y='price'))+
geom_point())
```

```python
(ggplot(data=mpg, mapping=aes(x='cty', y='hwy'))+
geom_point())
```
